```python
from bs4 import BeautifulSoup
import requests
```

# Scrap HackerNews Home page

```python
response = requests.get("https://news.ycombinator.com/")
response.raise_for_status()

soup = BeautifulSoup(response.content, 'html.parser')
# print(soup.prettify())
```

```python
titleslink = soup.select(".athing .title .titlelink")
scores = soup.select(".subtext")
```

```python
items = []
for i in range(len(titleslink)):
    title = titleslink[i].text
    title_url = titleslink[i].get('href')
    score = scores[i].select(".score")
    # Some artical might not have score
    if score:
        score = int(scores[i].text.split(" ")[0])
    else:
        score = 0
    items.append({'title': title, 'url': title_url, "score": score})

items = sorted(items, key=lambda x: x["score"], reverse=True)
items
```
